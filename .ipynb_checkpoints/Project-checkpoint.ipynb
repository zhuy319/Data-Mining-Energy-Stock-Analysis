## Clustering

```python
# define functions to prepare animal data for clustering
def adjust_feature_types(dataset):
    # convert numeric features into float64
    numeric_columns = dataset.select_dtypes('number').columns
    for column in numeric_columns:
        dataset = dataset.astype({column: 'float64'})
    # convert object features into categorical
    obj_cols = dataset.select_dtypes('object').columns
    for column in obj_cols:
        dataset = dataset.astype({column: 'category'})
    # apply one hot encoding for animal species and commodity
    dataset = my_utils.encode_features(dataset)     
    return dataset
```

```python
# prepare, clean, and do scaling
import my_utils
from sklearn.decomposition import PCA

def prepare_data(dataset):
    dataset = adjust_feature_types(dataset)
    
    return dataset

def clean_data(data):
    origlen = data.shape[0]
    print("Original number of instances is ", origlen)
    data = my_utils.remove_outliers(data, num_std=5)
    print("Number of removed outliers further than 5 standard deviation from the mean is ",(origlen-data.shape[0]))
    data = data.drop_duplicates()
    print("Number of removed duplicated instances is ",(origlen-data.shape[0]))
    data = my_utils.remove_miss_columns(data)
    print("Number of values with NA being interpolated is ",data.isna().sum().sum())
    data = data.interpolate('linear')
    return data
    
class Preprocessor:
    # initialize constructors
    def __init__(self):
    # For instance operations
        # parameter for remove outliers
        self.feature_means = None
        self.feature_std = None
    
    # For feature operations
        # parameter for selected features
        self.selected_features = None
        # parameter for kselected features
        self.kselected_features = None
    
    # For value operations
        # parameter for scale values
        self.scaler = preprocessing.StandardScaler()
        self.pca = PCA(n_components = 2)
    
    # fit dataset and store constructor objects
    def fit(self, data, X_clf, Y_clf, k, flags):
        #self.selected_features = my_utils.basic_features_selection(data).columns
        #numeric_columns = data[self.selected_features].select_dtypes('float64').columns
        self.selected_features = data.columns
        numeric_columns = data.select_dtypes('float64').columns
        self.scaler = self.scaler.fit(data[numeric_columns])
        if flags[0]==1:
            self.kselected_features = my_utils.k_features_selection(data,X_clf,Y_clf,k).columns
        if flags[1]==1:
            self.pca = self.pca.fit(data[self.selected_features])
        
    # transform dataset with flags 
    def transform(self, data, flags):
        data = data[self.selected_features]
        numeric_columns = data.select_dtypes('float64').columns
        data = data.copy()
        data[numeric_columns] = self.scaler.transform(data[numeric_columns])
        kbestdata = pd.DataFrame()
        pcadata = pd.DataFrame()
        if flags[0]==1:
            kbestdata = data[self.kselected_features]
        if flags[1]==1:
            pcadata = self.pca.transform(data)
        return data, kbestdata, pcadata
```

```python
dataset = prepare_data(df2)
print(dataset.columns)
flags = [0,1]

# do basic cleaning on the dataset
dataset = clean_data(dataset)

# instantiate Preprocessor class
preprocessor = Preprocessor()

# fit the dataset and transform
preprocessor.fit(dataset,None,None,0,flags)
dataset, kbestdata, pcadata = preprocessor.transform(dataset,flags)
```

We thus pick high price in each day, and take average for the past 5 days, 30 days, and 60 days.

```python
# average high price for past 5 days
newdf = pd.DataFrame()
for i in range(1,19):
    newdf.at[i-1,0]=dataset.iloc[0:4,(i-1)*6].mean()
```

## Clustering Conclusion
Generally speaking, there is no strong preference for renewable/nonrenewable stocks belonging to a specific cluster.

We can see that even though cluster 2 only includes one stock WAVE, which seems to be too imbalanced. It is very unlike other stocks since it has a much higher 5-day average, indicating a recent increase in its stock price. This large increase can also be shown from the difference between the 60 days, 30 days, and 5 days average, which offers a significant growth trend. Also, the 30-day and 60-day average also slightly higher than other clusters' average levels. 

Companies in cluster 1---including ORA, MNTK, VST, CWEN, ROCC, ORA, and USAC---have relatively stable and low 5 days, 30 days, and 60 days averages, which are all below zero. While companies in cluster 0--- AES, AMPS, VVPR, NOVA, AY, RNW, AZRE, ELLO, NRG, and THR---have relatively stable and higher 5 days, 30 days, and 60 days averag compared to companies in cluster 1. And there is a trend of a slight increase in these two months' stock price. 

Since stocks in cluster 0 only steadily increase that much in price, we may consider long them in the future. And stocks in cluster 1 have had stable low prices below 0 in the recent two months, so we may not long them in the short run, but their potential of getting more increase in price is more significant than other stocks because their price is lower. We may choose to short WAVE, which is in cluster 2, because it recently experienced a rapid increase, and it will likely fall back to its price, for example, 10 days ago, in the short run.

# Prediction Task

Firstly, we are going to predict the future stock price for each of the companies. Due to the nature of the time series, we can't randomize our train-test set which could cause a look-ahead bias and is not good for predicting the stock market. Instead, we decide to perform both blocking time series split and normal time series split for the cross-validation, and we'd like to compare the results for different splitting. In the model, we can adjust how many dates to perform our analysis, which makes our prediction model more flexible, and we can also compare the metrics of the model between the different amounts of input. We choose to run LSTM, decision tree, random forest, and linear regression and decide which has the best performance. Since it is an estimation task, we'd like to use mean absolute error (MAE) as our performance metric. Thus, after we choose the splitting method, the amount of input data, and the model with the best performance, we'd like to train our final model and make predictions on real data starting from 11-07-2022. We are going to make 20 days prediction and compare our prediction with the real stock price to see how accurate our predictions are and which stock to long or short based on our predictions.
