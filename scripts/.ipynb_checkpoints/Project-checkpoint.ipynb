```python
# define functions to prepare stock data for clustering
def adjust_feature_types(dataset):
    # convert numeric features into float64
    numeric_columns = dataset.select_dtypes('number').columns
    for column in numeric_columns:
        dataset = dataset.astype({column: 'float64'})
    # convert object features into categorical
    obj_cols = dataset.select_dtypes('object').columns
    for column in obj_cols:
        dataset = dataset.astype({column: 'category'})
    # apply one hot encoding 
    dataset = my_utils.encode_features(dataset)     
    return dataset
```

```python
# prepare, clean, and do scaling
import my_utils
from sklearn.decomposition import PCA

def prepare_data(dataset):
    dataset = adjust_feature_types(dataset)  
    return dataset

# we would not include remove outliers in this funtion because huge fluctuations in stock markets are good to investigate on.
# There may be some big events happening to whole market or specific sector or company.
# Also, since some stocks don't have data beginning in 2016, we just get rid of those time during which some stocks' data are lacking.
def clean_data(data):
    origlen = data.shape[0]
    print("Original number of instances is ", origlen)
    data = data.drop_duplicates()
    print("Number of removed duplicated instances is ",(origlen-data.shape[0]))
    data = data.dropna()
    print("Number of removed instances with NA is ",(origlen-data.shape[0]))
    return data
```

From the visualization of KMeans's and MiniBatchKMeans's clustering results with PCA, we can see that the clustering of KMeans seems to be more reasonable than that of MiniBatchKMeans. Because, for example, the point in the lower left corner (in cluster 0) of MiniBatchKMeans may be better if it is clustered into cluster 1 because it is closer to that cluster, or it can even be one cluster itself. Also, the point on the very right in cluster 2 in MiniBatchKMeans plot is clearly an outlier and should be in a single cluster itself.

Similarly to the results from visualization of the results of KMeans and MiniBatchKMeans, the silhouette scores for KMeans are very good because there are almost no negative silhouette coefficient values which means there is little incorrect clustering. From all the visualizations and evaluations, we can get the conclusion that KMeans works better for clustering our stock dataset, hence we would go with "cluster", which is the clustering results of KMeans.

We then tried new clustering method using hierarchial clustering. Comparing single link, complete link, and group average, if we draw a vertical line to cut the stocks into exactly 4 clusters, complete link is a better choice because single link gives a very huge cluster containing 49 stocks, and three small clusters containing 2,1,1 stocks respectively.
```python
def average_calculator(dataframe,value,labels):
    newdf = pd.DataFrame()
    for i in range(0,int((len(dataframe.index))/value)):
        for j in range(0,len(dataframe.columns)):
            newdf.at[i,j]=dataframe.iloc[(0+i*value):(value+i*value),j].mean()
    newdf.columns=labels      
    return newdf
```

```python
def association_preprocess(dataset,value,labels):
    # instantiate Preprocessor class
    association_preprocessor = preprocessor.Preprocessor(0)
    # fit the dataset and transform
    association_preprocessor.fit(dataset)
    dfA = association_preprocessor.transform(dataset)
    labels =labels
    dfA = pd.DataFrame(dfA, columns = labels)
    dfA=average_calculator(dfA,value,labels)
    transactions_list = list(dfA.apply(lambda x: list(dfA.columns[x>0]), axis=1))
    df = association_preprocessor.encoder(transactions_list)
    return df
```

#### Task 3: Prediction

In the estimation task, we are going to predict the future 15-day prices for each stock in cluster 0 which we found in our clustering analysis that they would have a better short-term return compared to stocks in other clusters. During the task, we will only  use dates as features and close price as our response variable because we assume that we won't have any additional information such as high price, low price, and so on for future stocks.

```python
prediction_train = [FSLR, FREY, LIN, GM, ETN, ORA, AES, OXY, TTE, CVX, KMI, PXD, OVV, KBR, CIVI, KNTK, ROCC, NRG, USAC]
prediction_names = ['FSLR', 'FREY', 'LIN', 'GM', 'ETN', 'ORA', 'AES', 'OXY', 'TTE', 'CVX', 'KMI', 'PXD', 'OVV', 'KBR', 'CIVI', 'KNTK', 'ROCC', 'NRG', 'USAC']
```

```python
# disctribution of target features
close_price = []
for i in range(len(prediction_train)):
    close_price.append(prediction_train[i]['Close'])
close_price = pd.DataFrame(close_price).transpose()
close_price.columns = prediction_names
plt.subplots(figsize = (16,9))
sns.boxplot(data = close_price);
```

We can see from the graph that each company has very differently-ranged close distributions, so it's needed for us to find the model for each stock individually.

```python
def prediction_data_preparation(data):
    data['Date'] = data.index
    data = data.sort_index(ascending=True,axis=0)
    new_dataset = data[['Date', 'Close']]
    new_dataset = new_dataset.reset_index(drop = True)
    new_dataset['Date'] = pd.to_datetime((new_dataset['Date']))
    new_dataset['Date_Year'] = new_dataset['Date'].dt.year
    new_dataset['Date_Month'] = new_dataset['Date'].dt.month
    new_dataset['Date_Day'] = new_dataset['Date'].dt.day
    new_dataset['Date_WeekDay'] = new_dataset['Date'].dt.dayofweek
    return new_dataset
```

To prepare the data for estimation, we performed feature engineering for 'Dates', which is the index of the stock datasets, and we derived numerical features of the year, month, day, and day of the week from dates. 

```python
def test_data_generation(start_date, end_date):
    test_data = pd.DataFrame(pd.date_range(start=start_date, end=end_date, freq='B'), columns=['Date'])
    test_data['Date'] = pd.to_datetime(test_data['Date'])
    test_data['Date_Year'] = test_data['Date'].dt.year
    test_data['Date_Month'] = test_data['Date'].dt.month
    test_data['Date_Day'] = test_data['Date'].dt.day
    test_data['Date_WeekDay'] = test_data['Date'].dt.dayofweek
    return test_data
```

Then we run the cross-validation of 4 models including linear regression, support vector machine, decision tree, and random forest, under the blocked time series split  which takes 20 instances as the fixed input, and the normal time series split which takes a growing number of instances as the input, for each stock in order to choose the best-performed combination of model and split approach.

```python
fittings = []
model_names = []
model_splits = []
best_models = []
prediction_preprocessors = []
for i in range(len(prediction_train)):
    print('Running cross validation for', prediction_names[i], '...')
    new = prediction_data_preparation(prediction_train[i])
    fitting, model_name, model_split, best_model, prediction_preprocessor = prediction.cross_validation(new)
    print('-----------------------------------------------------------------------------------')
    fittings.append(fitting)
    model_names.append(model_name)
    model_splits.append(model_split)
    best_models.append(best_model)
    prediction_preprocessors.append(prediction_preprocessor)
```

```python
fittings = pd.DataFrame(fittings)
fittings.columns = ['Naive', 'Training', 'Test']
fittings.index = prediction_names
fittings
```

During the cross-validation, we used mean absolute error as our performance metrics, chose the combo of model and approach with the smallest test MAE for each stock, and made a visualization of the naive training MAE, training MAE, and test MAE to check underfitting and overfitting.

From figure above, we can see that all the models have lower training MAEs than naive training MAEs, so we are fine with the problem of underfitting. However, overfitting is the problem that we are concerned about especially for the models of stocks LIN, PXD, and CIVI because their models show a much higher value of their test MAEs than training MAEs. This is probably because we only predict the stock prices using the barely-informational feature 'Dates', overfitting is very common under this scenario, and we can always try to solve the problem by decreasing the complexity of the models. However, in consideration of our limited computational expenses, we kept the overfitted model for other stocks and only try to simplify models for LIN, PXD, and CIVI.

```python
performances = []
for i in range(len(prediction_train)):
    act = prediction_test[i]['Close'].tail(15)
    pre = prediction_test[i]['Predictions'].tail(15)
    residual = act-pre
    MAE = mean_absolute_error(act, pre)
    max_error = max(residual)
    r2 = r2_score(act, pre, multioutput='uniform_average')
    act1 = prediction_test[i]['Close']
    pre1 = prediction_test[i]['Predictions']
    trend_correct = 0
    for j in range(0, prediction_test[i].shape[0]-1):
        if (act1[j+1]-act1[j])*(pre1[j+1]-pre1[j]) > 0:
            trend_correct += 1
    trend_accuracy = (trend_correct/15)*100
    performances.append([MAE, max_error, r2, trend_accuracy])
performances = pd.DataFrame(performances).round(decimals=2)
performances.index = prediction_names
performances.columns = ['Test MAE', 'Max Error', 'Test R2','Trend Accuracy(%)']
performances.to_csv('../output/performance.csv', index=True)
performances
```

Table above shows 4 different test metrics for our prediction. Thus, we calculated the R-squares and accuracy of the trend's consistency between the actual price and our prediction, and R-squares are negative ranging from -87.96 to -0.11, which is possible because we evaluated models separately on test data, and this indicates that the models are not very predictive. The accuracy of the trend's consistency varies from 6.67\% to 66.67\%, meaning that our predictions catch 6.67\% to 66.67\% of the actual trends. We are not very satisfied with our on-time prediction results, and several factors we didn't include such as the holiday of Thanksgiving might be responsible for the unsatisfying results, so we should try more independent datasets of varying time periods.
